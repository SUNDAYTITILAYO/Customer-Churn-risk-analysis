Loading & Inspecting Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#Loading the dataset
path = "/content/Telco-Customer-Churn.csv"
df = pd.read_csv(path)
df.head()
```

##EXPLORATORY DATA ANALYSIS (EDA)

```python
# showing the lists of the columns
df.columns.tolist()
```

```python
df.info()
```

From the info, Total Charges column is an object but it takes a numeric data, so we will need to convert it to a numeric column so that it doesn't affect us  later.


```python
# Changing the Total Charges column to a numeric column because it was originally an object dtype.
df["Total Charges"] = pd.to_numeric(df["Total Charges"], errors="coerce")

```

```python
df.isnull().sum()
```

The total Charges Column now reflects 11 missing values after converting it to a numeric column while other columns do not have any missing value.

```python
# Filling the missing values in total Charges Column.
df["Total Charges"].fillna(df["Total Charges"].median(), inplace=True)
```

```python
# checking to confirm all the missing values have been filled.
df["Total Charges"].isnull().sum()
```

All the missing Values have now being filled.

CONVERTING CHURN

```python
# converting the target (Churn)
df["Churn Label"]= df["Churn Label"].map({"Yes": 1, "No" : 0})
```

```python
# checking the Churn counts
df["Churn Label"].value_counts()
```

```python
df.columns
```

The above shows that the churn rate is 26% which means that;


*   Out of every **100** customers **26** leave



*   That's more than **1** in **4** **Customers**

*   loosing 1 in 4 customers shows that there is a ***major revenue leakage***






If the Churn rate is this high, it means there is a retention problem and we

need to know which customers are leaving and why.





#Revenue KPI

```python
avg_revenue = df["Monthly Charges"].mean()
total_revenue = df["Monthly Charges"].sum()

print(avg_revenue, total_revenue)
```

#Churn By Tenure Group

```python
# creating Tenure group
df["tenure_group"] = pd.cut(df["Tenure Months"],
bins = [0,12,24,48,72],
labels = ['0-1yr','1-2yr','2-4yr','4-6yr'])

#calculating churn by tenure group
df.groupby("tenure_group")["Churn"].mean()
```

**Insight:**


*  Customers in the **0–1 year** group have **~48%** churn (very high)
*  Customers in **4–6 years** have **~9.5%** churn (very low)


**Business Interpretation:**

*  Early-stage customers are at highest risk
*  Long-term customers are stable and valuable


**Recommendation:**


*  Improve onboarding and early retention strategies
*  Offer incentives for new customers

```python
# creating a copy for tableau dashboard
tableau_df = df.copy()
```

#MODELLING

```python

# Dropping CustomerID Column
model_df = df.drop(['CustomerID','Churn Score','Churn Reason','Churn Value','Churn','Churn Label','Country','State','City','Lat Long','Unnamed: 0'], axis = 1)

#Converting Categorical Columns to Numeric
model_df = pd.get_dummies(model_df, drop_first = True)

#Defining Features and Target
x = model_df
y = df['Churn']

#Splitting the Data
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state=42)

#Training the Model
Model = LogisticRegression(max_iter = 1000)
Model.fit(x_train,y_train)
Model.score(x_train,y_train)

#Creating Churn Risk Score
tableau_df["Risk_score"] = Model.predict_proba(x)[:, 1]
tableau_df["Risk_score"].head()

#previewing the risk score
tableau_df[['Risk_score','Churn']].head()

```

with the risk score above, we can now predict which customers are likely to churn and take a proactive measure before they leave.

##Creating Risk Segment

```python
tableau_df['Risk_segment']= pd.cut(tableau_df['Risk_score'], 3,
                    labels = ['Low','Medium','High'])
tableau_df[['Risk_score','Risk_segment']]
```
